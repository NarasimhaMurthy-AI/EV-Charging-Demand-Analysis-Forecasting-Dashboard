```python
# Time Series Analysis

Objective:
Analyze trends in EV charging demand using rolling averages.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(
    "data/processed/EV_Cleaned_Hourly.xlsx",
    index_col=0,
    parse_dates=True
)

df.head()
```

```python
# Rolling average
df["Rolling_24h"] = df["EV_Usage"].rolling(24).mean()

plt.figure(figsize=(12,5))
plt.plot(df["EV_Usage"], label="Original Demand", alpha=0.6)
plt.plot(df["Rolling_24h"], label="24-Hour Rolling Average", linewidth=2)
plt.legend()
plt.title("EV Charging Demand Trend")
plt.xlabel("DateTime")
plt.ylabel("kW")
plt.grid(True)
plt.show()
```

```python
Conclusion:
Rolling average smooths short-term fluctuations and highlights
overall demand trend, useful for grid stability planning.
```
