```python
# Exploratory Data Analysis (EDA)

Objective:
Analyze EV charging demand trends and usage patterns.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(
    "data/processed/EV_Cleaned_Hourly.xlsx",
    index_col=0,
    parse_dates=True
)

df.head()
```

```python
# Hourly demand trend
plt.figure(figsize=(12,5))
plt.plot(df["EV_Usage"])
plt.title("Hourly EV Charging Demand")
plt.xlabel("DateTime")
plt.ylabel("kW")
plt.grid(True)
plt.show()
```

```python
# Average demand by hour
df["Hour"] = df.index.hour

df.groupby("Hour")["EV_Usage"].mean().plot(
    kind="bar",
    figsize=(10,4),
    title="Average EV Charging Demand by Hour"
)
plt.xlabel("Hour of Day")
plt.ylabel("kW")
plt.show()
```

```python
Insights:
- Peak demand observed during specific daytime hours
- Lower usage during late-night hours
- Indicates opportunity for load balancing
```
